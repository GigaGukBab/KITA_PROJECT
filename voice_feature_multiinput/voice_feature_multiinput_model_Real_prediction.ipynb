## 지수_보이스피싱 데이터 예측

```python
import os
import numpy as np
from tensorflow.keras.models import load_model

# Pitch 데이터와 Decibel 데이터 파일 경로 설정
pitch_files_path = "./jisoo_vp_pitch"
decibel_files_path = "./jisoo_vp_dB"

# npy 파일 목록 가져오기
pitch_files = sorted([f for f in os.listdir(pitch_files_path) if f.endswith('.npy')])
decibel_files = sorted([f for f in os.listdir(decibel_files_path) if f.endswith('.npy')])

# 데이터 전처리 함수 작성
def preprocess_files(file_list, path):
    reshaped_files = []
    for file in file_list:
        data = np.load(path + '/' + file)
        reshaped_data = data.reshape(1, 157, 1)
        reshaped_files.append(reshaped_data)
    return np.vstack(reshaped_files).astype('float32')

# 사전에 저장된 모델 불러오기
model_best = load_model('./multi_input_model.h5')

# 모델 예측 함수 작성
def predict_with_model(model, pitch_files, decibel_files, pitch_path, decibel_path):
    pitch_data = preprocess_files(pitch_files, pitch_path)
    decibel_data = preprocess_files(decibel_files, decibel_path)
    probabilities = model.predict([pitch_data, decibel_data])
    labels = [1 if prob[0] >= 0.5 else 0 for prob in probabilities]
    
    for idx, probability in enumerate(probabilities):
        print(f"=== File {idx+1} ===")
        print(f"Class 0 Probability: {1 - probability[0]:.20f}")
        print(f"Class 1 Probability: {probability[0]:.20f}")
        print("Predicted Label:", labels[idx])
        print("=============================\n")
    
    return probabilities, labels

# 실제 파일로부터 데이터 로딩 및 예측 수행
print("\n=== Predictions ===\n")
predictions, labels = predict_with_model(model_best, pitch_files, decibel_files, pitch_files_path, decibel_files_path)
print("Predicted Labels List:", labels)
```
